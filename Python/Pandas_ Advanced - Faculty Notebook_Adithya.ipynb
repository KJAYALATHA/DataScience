## Pivot Tables
- It is used to display the data for the specified columns and index

```python
sales = {'Months': ['Jan','Feb','Mar','Apr','May','June','July','Aug','Sep','Oct','Nov','Dec'],
        'Sales': [22000,27000,25000,29000,35000,67000,78000,67000,56000,56000,89000,60000],
         'Seasons': ['Winter','Winter','Spring','Spring','Spring','Summer','Summer','Summer','Fall','Fall','Fall','Winter']
        }
```

```python
df_sales = pd.DataFrame(sales, columns= ['Months', 'Sales','Seasons'])
df_sales
```

```python
df_sales.info()
```

```python
df_sales.dtypes
```

```python
# create a pivot table where indexes are months
pd.pivot_table(df_sales,index=["Months"])
```

```python
# to create pivot table having multiple indexes
# index is a list of column names to be considered as indexes
pd.pivot_table(df_sales,index=["Months", "Seasons"])
```

```python
# create a pivot table for with the index as 'Seasons' and values as 'Sales'
# index is a list of column names to be considered as indexes
# values represent the numeric data
pd.pivot_table(df_sales,index=["Seasons"],values=["Sales"])
```

```python
#find the sum of sales in each season.
pd.pivot_table(df_sales,index=["Seasons"],values=["Sales"], aggfunc=np.sum)
```

```python
#find the sum of sales in each season.
pd.pivot_table(df_sales,index=["Seasons"],values=["Sales"], aggfunc=np.min)
```

```python
#find the sum of sales in each season.
pd.pivot_table(df_sales,index=["Seasons"],values=["Sales"], aggfunc=np.max)
```

```python
pd.pivot_table(df_sales,index="Months",values="Sales")
```

```python
pd.pivot_table(df_sales,index=["Seasons"],values=["Sales"], aggfunc=np.min)
```

```python
pd.pivot_table(df_sales,index=["Seasons"],columns="Months",values=["Sales"], aggfunc=np.min)
```

### Cross Tables
- Cross tables are similar to pivot tables
- It computes a cross tabulation of two or more factors
- It represent the frequency of occurrence of certain groups in the data.

```python
# read the csv file 
df_people = pd.read_csv('people.csv')

# print head of the data
df_people.head()
```

```python
df_people
```

```python
# find the count of of qualifications people have from each city
pd.crosstab(df_people.Qualification, df_people.Hometown)
```

```python
df_people.columns
```

```python
pd.crosstab(df_people.Gender, df_people.Smoker)
```

```python
pd.crosstab(df_people.Gender, df_people.Hometown)
```

```python
# find the average age of smokers based on their city of residence
# values represent the numeric data
# aggfunc specifies the function
pd.crosstab(df_people.Hometown, df_people.Smoker, values = df_people.Age, aggfunc = 'mean')
```

```python
pd.crosstab(df_people.Hometown, df_people.Smoker, values = df_people.Age, aggfunc = 'min')
```

```python
pd.crosstab(df_people.Hometown, df_people.Smoker, values = df_people.Age, aggfunc = 'max')
```

```python
pd.crosstab(df_people.Hometown, df_people.Smoker, values = df_people.Age)
```

```python
pd.crosstab(df_people.Hometown, df_people.Smoker,aggfunc = 'max')
```

### Conclusion from cross tables
- values cannot be considered without aggregate function and vise-versa
- if values and aggregate function is not given, the output will give the count(frequency)

```python
pd.crosstab(df_people.Hometown, df_people.Smoker)
```

```python
pd.crosstab(df_people.Qualification, df_people.Hometown, rownames = ['Highest Qualification'], colnames=['City'])
```

```python
pd.crosstab(df_people.Qualification, df_people.Hometown)
```

```python
pd.crosstab(df_people.Hometown, df_people.Smoker, values = df_people.Age, aggfunc = 'mean')
```

### Checking the skewness
- Skewness is a lack of symmetry or departure from symmetry. The variable can be positively-skewed, negatively-skewed or no skewed.
- If the distribution of the data is elongated on either sides then the data is said to
be skewed
- If the distribution of the data is elongated on the left side then the data is said to
be left skewed
- If the distribution of the data is elongated on the right side then the data is said
to be right skewed

```python
(df_products['Sales']).skew()
#The value of skewness is near to zero; thus, we can interpret that the variable 'Sales' is not skewed.
```

```python
df_products['Profit'].skew()
```

```python
df_products.skew()
```

## Kurtosis
- It is a statistical measure that defines how heavily the tails of the distribution
differ from the normal distribution
- It identifies whether the tails of a given distribution contain extreme values
- Thicker the tails are more the extreme values in the data

```python
(df_products['Profit']).kurt()
#Here, the value of kurtosis is negative, which implies that 'Profit' is a platykurtic distribution
```
