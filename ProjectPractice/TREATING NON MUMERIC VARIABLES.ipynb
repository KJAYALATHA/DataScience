```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('bank-additional-full.csv',sep=';')
df.head(2)
```

### Job

```python
df['job'].value_counts()
```

```python
df.job.unique()
```

```python
# import the LabelEncoder
from sklearn.preprocessing import LabelEncoder

# instantiate the encoder
labelencoder = LabelEncoder()

# fit the encoder on 'Outlet_Size' 
df['Label_Encoded_Job'] = labelencoder.fit_transform(df.job)

# display first 5 observations
df.head()
```

## marital

```python
df.marital.value_counts()
```

```python
# import the OneHotEncoder
from sklearn.preprocessing import OneHotEncoder

# creating instance of one-hot-encoder
encode = OneHotEncoder()

# fit_transform: It returns a sparse array
# .toarray(): It returns a numpy array
df_encode = pd.DataFrame(encode.fit_transform(df[['marital']]).toarray(), columns = ['married_new', 'single_new', 
                                                                                         'divorced_new', 'marital_unknown_new'])



# merge with main dataframe (df_sales)
df_encode = pd.concat([df, df_encode], axis=1)

# print 20 rows of the data
df_encode.head(5)


```

# education

```python
df.education.value_counts()
```

```python
# import the OrdinalEncoder
from sklearn.preprocessing import OrdinalEncoder

# instantiate the encoder with the list of categories in the required order
orderencoding = OrdinalEncoder(categories = [["illiterate", "basic.4y", "basic.6y", "basic.9y","high.school","university.degree","professional.course","unknown"]])

# add a column of ordered labels
# reshape(-1, 1): to rearrange the data
df_encode['education_encoded_new'] = orderencoding.fit_transform(df_encode['education'].values.reshape(-1,1))

# display the data
df_encode.head()
```
